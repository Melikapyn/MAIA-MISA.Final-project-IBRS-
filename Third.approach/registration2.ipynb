Registration of atlas 9 to validation set

```python
import subprocess
import os

# Define paths
elastix_path = '/Users/melikapooyan/Documents/elastix-5.0/elastix'  # Path to the Elastix executable
fixed_image_paths = [
    '/Users/melikapooyan/Documents/elastix-5.0/Misa.course/IBSR/Validation_Set/IBSR_11/IBSR_11_preprocessed_histmatched.nii.gz',
    '/Users/melikapooyan/Documents/elastix-5.0/Misa.course/IBSR/Validation_Set/IBSR_12/IBSR_12_preprocessed_histmatched.nii.gz',
    '/Users/melikapooyan/Documents/elastix-5.0/Misa.course/IBSR/Validation_Set/IBSR_13/IBSR_13_preprocessed_histmatched.nii.gz',
    '/Users/melikapooyan/Documents/elastix-5.0/Misa.course/IBSR/Validation_Set/IBSR_14/IBSR_14_preprocessed_histmatched.nii.gz',
    '/Users/melikapooyan/Documents/elastix-5.0/Misa.course/IBSR/Validation_Set/IBSR_17/IBSR_17_preprocessed_histmatched.nii.gz'
]
moving_image_path = '/Users/melikapooyan/Documents/elastix-5.0/mean_9.nii.gz'  # Path to the moving image

param_file_path1 = '/Users/melikapooyan/Documents/elastix-5.0/Misa.course/Parameters.Par0009.affine.txt'  # Path to the first parameter file
param_file_path2 = '/Users/melikapooyan/Documents/elastix-5.0/Misa.course/Parameters.Par0009.elastic.txt'  # Path to the second parameter file

output_dir_base = '/Users/melikapooyan/Documents/elastix-5.0/atlas9'  # Base output directory

for fixed_image_path in fixed_image_paths:
    # Extract the identifier of the fixed image for the output directory
    fixed_image_identifier = os.path.basename(fixed_image_path).split('.')[0].split('_')[1]
    output_dir = os.path.join(output_dir_base, f'resIBSR_{fixed_image_identifier}')

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Construct the Elastix command with two parameter files
    command = [
        elastix_path,
        '-f', fixed_image_path,
        '-m', moving_image_path,
        '-out', output_dir,
        '-p', param_file_path1,
        '-p', param_file_path2
    ]

    # Execute the Elastix command
    result = subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

    # Print the output for debugging
    print(f"Registration with fixed image IBSR_{fixed_image_identifier}")
    print(result.stdout)

    # Check if Elastix ran successfully
    if result.returncode == 0:
        print(f'Elastix registration completed successfully for IBSR_{fixed_image_identifier}.')
    else:
        print(f'Elastix registration failed for IBSR_{fixed_image_identifier}.')
        print(result.stderr)
```

```python
import subprocess
import os

def run_transformix(input_image_path, transform_param_file, output_dir):
    # Ensure the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Construct the Transformix command
    command = [
        transformix_path,
        '-in', input_image_path,
        '-out', output_dir,
        '-tp', transform_param_file
    ]

    # Execute the Transformix command
    result = subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

    # Print both stdout and stderr for diagnostics
    print("Standard Output:", result.stdout)
    print("Error Output:", result.stderr)

    # Check if Transformix ran successfully
    if result.returncode == 0:
        print(f'Transformix completed successfully for {input_image_path}.')
    else:
        print(f'Transformix failed for {input_image_path}.')

# Define paths
transformix_path = '/Users/melikapooyan/Documents/elastix-5.0/transformix'
wm_map_path = '/Users/melikapooyan/Documents/elastix-5.0/wm_1.nii.gz'
transform_param_file = '/Users/melikapooyan/Documents/elastix-5.0/atlas9/resIBSR_17/TransformParameters.1.txt'
output_dir_wm = '/Users/melikapooyan/Documents/elastix-5.0/atlas9/resIBSR_17/wm'

# Run Transformix for the CSF map
run_transformix(wm_map_path, transform_param_file, output_dir_wm)
```
